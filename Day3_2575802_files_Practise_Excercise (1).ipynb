# 1. Write a python function that copies a file reading and writing up to 50 characters at a time

```python
def copyfiles(src_file, dest_file):
    with open(src_file, 'rb') as source:
        with open(dest_file, 'wb') as destination:
            while True:
                data = source.read(50)
                if not data:
                    break
                destination.write(data)
    print(f"File '{src_file}' copied to '{dest_file}' successfully.")


sourceFile = "source.txt"
source= open(sourceFile, "w")
source.write("Hello, World! I am Raja\nTo top that off, Abu Zubaydah's lawyers are prohibited from revealing any contents of any conversations they have had with their client, on the grounds that any disclosures could threaten national security.")
source.write("This is a sample file created with Python.\nYou can use it as an indefinite pronoun or to refer to any kind of material, such as the contents of a book, email, or blog post. For example: Please read through the book and note down any contents that you find particularly interesting.")

source.close()

destinationFile= "destination.txt"
destination=open(destinationFile,"w")
destination.close()

copyfiles(sourceFile, destinationFile)
```

# 2. Print all numbers present in the textfile and print the number of blank spaces in that file

```python
def numbers_blanksCount(file_path):
        with open(file_path, 'r') as file:
            content = file.read()
            numbers = []
            blank= 0

            for char in content:
                if char.isdigit():
                    num = char
                    while True:
                        next_char = file.read(1)
                        if next_char.isdigit() or next_char == '.':
                            num += next_char
                        else:
                            break
                    numbers.append(num)
                elif char.isspace():
                    blank += 1

            if numbers:
                print("Numbers in the file:", numbers)
            else:
                print("No numbers found in the file.")
            print("Number of blank spaces in the file:", blank)


File = "source.txt" 
numbers_blanksCount(File)
```

# 3. Write a function called sed that takes as arguments a pattern string, a replacement string, and two filenames; it should read the first file and write the contents into the second file(creating it if necessary). If the pattern string appears anywhere in the file, it should be replaced with the replacement  string

```python
def sed(pattern, replacement, input_file, output_file):
    try:
        with open(input_file, 'r') as file:
            content = file.read()
            new_content = content.replace(pattern, replacement)

        with open(output_file, 'w') as output:
            output.write(new_content)

        print(f"Pattern '{pattern}' replaced with '{replacement}' "
              f"and saved to '{output_file}' successfully.")
    except Exception as e:
        print(f"An error occurred: {e}")

pattern = "Python"        
replacement = "Java"       
input_file = "example.txt" 

file_name = "output.txt"
file = open(file_name, "w")
file.close()

output_file = "output.txt"     

sed(pattern, replacement, input_file, output_file)
```

# 5.You have a text file with a large amount of text, and you want to search for specific words or phrases and replace text them with new content

```python
def search_replace(file_path, search_text, replace_text):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    with open(file_path, 'w') as file:
        for line in lines:
            modified_line = line.replace(search_text, replace_text)
            file.write(modified_line)

    print(f"Text replaced successfully in '{file_path}'.")
        
        
file_path = "output.txt"
search = "java"  
replace = "C sharp"  

search_replace(file_path, search, replace)
```

# 6.Write a python script that concatenates the content of multiple text files into a single output file. Allow the user to specify the input files and output file

```python
def merge_file(inputfile, outputfile):
    with open(outputfile, 'w') as output:
        for input_file in inputfile:
            with open(input_file, 'r') as file:
                content = file.read()
                output.write(content)
            print(f"Concatenated '{input_file}' to '{outputfile}'")

    print(f"Concatenation completed. Merged content saved to '{outputfile}'")


inputfile = ["example.txt", "destination.txt"]
file_name = "merge.txt"
file = open(file_name, "w")
file.close()
outputfile = "merge.txt"
merge_file(inputfile, outputfile)
```

# 8. You are developing a student gradebook system for a school.

1.The system should allow teachers to input student grades for various subjects, 2.store the data in seperate text files for each subject
 3. provide students with the ability to view their grades.
Using python
4.With error handling

```python
gradebook = {}

def input_grades(subject):
    grades = {}
    while True:
        student_id = input("Enter student ID (or 'q' to quit): ")
        if student_id == 'q':
            break
        try:
            grade = float(input(f"Enter grade for Student ID {student_id}: "))
            grades[student_id] = grade
        except ValueError:
            print("Invalid input. Please enter a valid grade in number.")
    gradebook[subject] = grades


    with open(f"{subject}.txt", "w") as file:
        for student_id, grade in grades.items():
            file.write(f"{student_id}: {grade}\n")
    
    print(f"Grades for {subject} have been recorded in '{subject}.txt'.")

def view_grades(subject, student_id):
    if subject in gradebook:
        grades = gradebook[subject]
        if student_id in grades:
            print(f"Your grade for {subject}: {grades[student_id]}")
        else:
            print(f"No grade found for {student_id} in {subject}.")
    else:
        print(f"Grades for {subject} have not been recorded yet.")

while True:
    print("\nGradebook System")
    print("1. Teacher to Input Grades")
    print("2. Student to  View Grades")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        subject = input("Enter the subject name: ")
        input_grades(subject)
    elif choice == '2':
        student_id = input("Enter your student ID: ")
        subject = input("Enter the subject name: ")
        view_grades(subject, student_id)
    elif choice == '3':
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
