```python
roles_1950s = role[(role['year'] >= 1950) & (role['year'] <= 1959)]

actors_1950s = roles_1950s[roles_1950s['type'] == 'actor']
actresses_1950s = roles_1950s[roles_1950s['type'] == 'actress']

leading_roles_actors = actors_1950s[actors_1950s['n'] == 1].groupby('year').size()
leading_roles_actresses = actresses_1950s[actresses_1950s['n'] == 1].groupby('year').size()

print("Leading roles (n=1) available to actors in each year of the 1950s:")
print(leading_roles_actors)

print("\nLeading roles (n=1) available to actresses in each year of the 1950s:")
print(leading_roles_actresses)
```

```python
#In the 1950s decade taken as a whole, how many total roles were available to actors, and how many to actresses, for each "n" number 1 through 5?
```

```python
roles_1950s = role[(role['year'] >= 1950) & (role['year'] <= 1959)]

actors_1950s = roles_1950s[roles_1950s['type'] == 'actor']
actresses_1950s = roles_1950s[roles_1950s['type'] == 'actress']

roles_count_by_n = {1: [], 2: [], 3: [], 4: [], 5: []}

for n in range(1, 6):
    roles_count_actors = actors_1950s[actors_1950s['n'] == n].shape[0]
    roles_count_actresses = actresses_1950s[actresses_1950s['n'] == n].shape[0]
    
    roles_count_by_n[n].append(roles_count_actors)
    roles_count_by_n[n].append(roles_count_actresses)

for n, counts in roles_count_by_n.items():
    print(f"n={n}: total roles for the Actors = {counts[0]},Total roles for the Actresses = {counts[1]}")
```
